map 함수
- map(매개변수(함수),매개변수(반복 가능한 자료형))
- map(적용시킬 함수, 적용할 값들)

```python
a = [1.1,2.2,3.3,4.4]
print(a)
```

```python
for i in range(len(a)):
    a[i] = int(a[i])
print(a)
```

```python
a = [1.1,2.2,3.3,4.4]
a = list(map(int,a))
print(a)
```

```python
# 리스트와 map을 이용하여 여러 값을 만들 수 있음
list(map(lambda x: x +100,[10,20,30]))
```

짝수의 값을 float형태로 바꾸기 (람다 형태로)

```python
a = [1,2,3,4,5,6]
```

```python
list(map(lambda x: float(x) if x % 2 ==0 else x,a))
```

```python
def float_x(x):
    if x % 2 ==0:
        return float(x)
    else:
        return x
result = list(map(float_x,a))
print(result)
```

- lambda 2개의 조건문 사용
- lambda의 매개변수들 : 식1 if 조건식1 else 식2 if 조건식2 else 식3 등으로 확장

- 값이 3인 경우 문자열로 변환, 짝수의 값을 float형태로 변경, 나머지 값은 *100 형태로 출력하라 (람다 형태로)

```python
list(map(lambda x: str(x) if x == 3 else float(x) if x % 2 ==0 else x*100,a))
```

```python
def abc(x):
    if x == 3:
        return str(x)
    elif x % 2 ==0:
        return float(x)
    else:
        return x * 100
list(map(abc,a))
```

```python
a = [1,2,3,4,5]
b = [6,7,8,9,10]
c = [0,0,0,0,0]
```

```python
# 매개변수가 2개 이상이면, 식의 변수도 2개 이상
list(map(lambda x,y,z :(x+y)*z, a,b,c))
```

- sort_values()

```python
df.sort_values('english')
# 오름차순이 디폴트
```

```python
df.sort_values('english',ascending=False)
# 내림차순으로 변환할 때는 ascending=False 사용
```

```python
df.sort_values('english',ascending=True)
```

```python
df.sort_values('english',ascending=True)[['id','science']].query('science >=60')
# 영어를 오름차순으로 정렬하고 그 데이터로 id와 science의 열만 출력하고 science가 60점 이상인 경우만 출력
```

오름차순, 내림차순 기준이 두 개 이상인 조건으로 차순 정리

```python
# 2 개 이상의 조건인 경우 2차원의 개념이라 리스트로 넣어야함
df.sort_values(['nclass','science'])
# nclass가 가장 앞에 있어서 메인, 서브가 science
```

```python
df.sort_values(['nclass','science'], ascending=[True, False])
# nclass에 True가 들어가 오름차순을 하고 서브로 science에 False 넣어 내림차순을 함
```
